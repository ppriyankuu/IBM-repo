```
import pandas as pd
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
```

```
wine = load_wine()
x = pd.DataFrame(wine.data, columns = wine.feature_names)
```

```
x
```

```
x.info()
```

```
x.duplicated()
```

```
x.describe().sum()
```

```
y = wine.data[:, 10]
y
```

```
x.head()
```

```
print('hue values for the first 5 rows', y[:5])
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
```

```
model = LinearRegression()
```

```
model.fit(x_train, y_train)
```

```
y_pred = model.predict(x_test)
```

```
from sklearn.metrics import mean_squared_error, r2_score
```

```
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```
print("Model cofficient:", model.coef_)
print("Model intercept:", model.intercept_)
print("Mean squared error:", mse)
print("R2 Score:", r2)
```

```
plt.scatter(y_test, y_pred, color='blue', alpha=0.7, label='Amit Karmakar')
plt.xlabel("Actual Hue Values")
plt.ylabel("Predicted Hue Values")
plt.title("Actual vs Predicted Hue Values")
plt.legend()
plt.show()
```
