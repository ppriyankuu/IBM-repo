```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/lung_cancer_prediction_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean() * 100
```

```python
df.drop(columns = ['Cancer_Stage', 'Treatment_Type'], inplace = True)
```

```python
df.isnull().sum().sum()
```

```python
diagnosed_cases = df[df['Lung_Cancer_Diagnosis'] == 'Yes']['Country'].value_counts()
top_10_countries = diagnosed_cases.sort_values(ascending = False).head(10)
# print(top_10_countries.index)
plt.bar(top_10_countries.index, top_10_countries.values, color = 'orange', edgecolor = 'black')
plt.title('Top 10 countries')
plt.xlabel('Countries')
plt.ylabel('counts')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

df.info()
```

```python
x = df.drop(columns = ['ID', 'Country', 'Population_Size', 'Age', 'Gender', 'Lung_Cancer_Diagnosis', 'Adenocarcinoma_Type', 'Annual_Lung_Cancer_Deaths'])
y = df['Lung_Cancer_Diagnosis']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB_model = GaussianNB()
```

```python
NB_model.fit(x_train, y_train)
```

```python
y_pred = NB_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f'NB Accuracy: {accuracy_score(y_test, y_pred):.4}')
```
